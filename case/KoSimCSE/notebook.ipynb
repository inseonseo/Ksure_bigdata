```python
def create_case_summary(data):
    """사건별로 모든 판정을 묶어서 요약 정보 생성"""
    
    case_summary = data.groupby(['보상파일번호', '사고번호']).agg({
        # 기본 사고 정보 (첫 번째 값)
        '사고접수일자': 'first',
        '사고금액': 'first',
        '사고유형': 'first',
        '수입국': 'first',
        '사고설명': 'first',
        
        # 판정 과정 정보 (모든 판정 포함)
        '판정일': list,
        '판정구분': list,
        '판정사유': list,
        '판정금액': list,
        '판정진행상태': list
    }).reset_index()
    
    return case_summary
```

```python
# 1. _x, _y로 끝나는 컬럼 이름 쌍 추출
columns_x = [col for col in test.columns if col.endswith('_x')]
columns_y = [col for col in test.columns if col.endswith('_y')]

# 매칭되는 prefix만 추려냄
common_cols = [col[:-2] for col in columns_x if col[:-2] in [c[:-2] for c in columns_y]]

# 2. 서로 다른 행을 모두 모아놓을 리스트
diff_rows = []

# 3. 각 컬럼 쌍마다 값이 다른 행 필터링
for col in common_cols:
    col_x = col + '_x'
    col_y = col + '_y'
    diff = test[test[col_x] != test[col_y]]
    if not diff.empty:
        print(f'🔍 {col}: 서로 다른 값이 존재합니다. (총 {len(diff)}건)')
        diff_rows.append(diff[['보상파일번호', '사고번호', col_x, col_y]].drop_duplicates())

# 4. 모든 차이나는 행 concat해서 보기 (선택)
if diff_rows:
    import pandas as pd
    all_diffs = pd.concat(diff_rows).drop_duplicates()
    display(all_diffs)
else:
    print("✅ 모든 _x/_y 컬럼 값이 동일합니다.")
```
