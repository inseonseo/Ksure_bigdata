## 1. Data Load, import library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df = pd.read_csv('Data/case.csv', encoding='cp949', index_col=0)
country = pd.read_csv('Data/country.csv', encoding='utf-8', index_col=0)
```

#### 1.1 컬럼 속성 변경 및 국가코드 merge

```python
df['수출자코드'] =df['수출자코드'].astype(str)
df['수입자코드'] = df['수입자코드'].fillna('').astype(str).str.replace(r'\.0$', '', regex=True)
```

```python
df['수출자코드'] = df['수출자코드'].str.zfill(3)
df['수입자코드'] = df['수입자코드'].str.zfill(3)
```

```python
country.reset_index(inplace=True)
# df.merge(country['수입자국가코드'])
```

```python
country.sample(1)
```

```python
country['수입자코드'] = country['수입자국가코드']
country['수출자코드'] = country['수입자국가코드']
```

```python
df.shape, df.merge(country, on='수출자코드').shape, df.merge(country, on='수입자코드').shape
```

```python
df = df.merge(country[['수출자코드', '국가한글명']], on='수출자코드', how='left')
df.rename(columns={'국가한글명' : '수출자국가'}, inplace=True)
df = df.merge(country[['수입자코드', '국가한글명']], on='수입자코드', how='left')
df.rename(columns={'국가한글명' : '수입자국가'}, inplace=True)
df.sample(1)
```

```python
df.drop(columns=['수출자코드', '수입자코드'], inplace=True)
```

```python
df.to_csv('Data/case.csv', encoding='cp949')
```
